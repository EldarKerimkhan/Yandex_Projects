**Технологический процесс**
- Rougher feed — исходное сырье,
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant

- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

**Параметры этапов**
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

### Рассчет эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно.

Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдем MAE между расчётами и значением признака.

Эффективность обогащения рассчитывается по формуле



где:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

```python
C = data_train['rougher.output.concentrate_au'] 
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']

data_train['rougher.output.recovery_new'] = ((C * (F - T)) / (F * (C - T))) * 100
# Выведем рассчитанный коэффициент обогащения и реальный
data_train[['rougher.output.recovery_new', 'rougher.output.recovery']]
```

Найдем MAE между расчётами и значением признака.

```python
mae = mean_absolute_error(data_train['rougher.output.recovery'], 
                          data_train['rougher.output.recovery_new'])
mae
```

Так как метрика MAE между рассчитанным нами коэффициентом и значением признака имеет очень маленькое значение, то можем сделать вывод, что эффективность обогащения рассчитана правильно.

## Анализ данных

**Изменение концентрации металлов (Au, Ag, Pb) на различных этапах очистки.**

Выделим столбцы с наличием металла в названии и слова '.concentrate' или '.feed' (для дальнейшего анализа размера гранул сырья).

```python
metals = ['au', 'ag', 'pb']
stages = ['rougher.input', 'rougher.output','primary_cleaner', 'final']

metal_columns = []
for column in data_full.columns:
    for metal in metals:
        if (metal in column and '.concentrate' in column) or (metal in column and '.feed' in column):
            metal_columns.append(column)

metal_columns
```

Сгрупперуем таблицу по названию металла. Также добавим столбец этап и среднюю концентрацию металла на данном этапе.

Создадим функцию, которая вернет нам новый столбец с названиями металлов

```python
def rename_metal(metal_columns):
    new_metal_column = []
    for column in metal_columns:
        for i in metals:
            if i in column:
                new_metal_column.append(i)
    return new_metal_column   
```

```python
def rename_stage(metal_columns):
    new_stage_column = []
    for column in metal_columns:
        for i in stages:
            if i in column:
                new_stage_column.append(i)
    return new_stage_column   
```

```python
metal_columns = ['rougher.input.feed_ag',
 'rougher.input.feed_pb',
 'rougher.input.feed_au',
 'rougher.output.concentrate_ag',
 'rougher.output.concentrate_pb',
 'rougher.output.concentrate_au',
 'primary_cleaner.output.concentrate_ag',
 'primary_cleaner.output.concentrate_pb',
 'primary_cleaner.output.concentrate_au',
 'final.output.concentrate_ag',
 'final.output.concentrate_pb',
 'final.output.concentrate_au']

data_plot = pd.DataFrame(data_full[metal_columns].median().reset_index()
                         .set_axis(['index', 'concentration'], axis=1, inplace=False))
data_plot = data_plot.drop('index', axis=1)
data_plot['metal'] = rename_metal(metal_columns)
data_plot['stage'] = rename_stage(metal_columns)
```

```python
data_plot
```

```python
ax = sns.catplot(data=data_plot, kind='bar', x='stage', y='concentration', 
                 col='metal', height=6, aspect=1)
ax.set_xticklabels(rotation=45, fontsize=15)
```

**Вывод:**

1. Из графиков видно, что концентрация золота увеличивается от этапа к этапу. На выходе концентрация в 4 раза больше, чем в исходном сырье.

## Модель

**Напишем функцию для вычисления итоговой sMAPE.**

```python
# рассчет метрики sMAPE
def smape_calc(target, predict):
    
    numerator = np.abs(target - predict)
    denominator = (np.abs(target) + np.abs(predict)) / 2 

    return 1 / len(target) * np.sum(numerator / denominator) * 100 
```

```python
# рассчет итоговой метрики sMAPE
def total_smape(rougher_smape, final_smape):
    total = 0.25 * rougher_smape + 0.75 * final_smape
    return total
```

```python
# создадим из функции sMAPE метрику качества
smape = make_scorer(smape_calc, greater_is_better=False)
```

```python
def objective(trial, features_train, target_train, features_test, target_test):
    params = {
        'iterations' : trial.suggest_int('iterations', 100,1000),
        'learning_rate' : trial.suggest_float('learning_rate', 0.01, 1),
        'depth' : trial.suggest_int('depth', 1,15)
    }
    train_rougher_model = CatBoostRegressor(random_state=123, **params, verbose=False)
    train_rougher_model.fit(features_train, target_train)
    prediction = train_rougher_model.predict(features_test)
    return smape_calc(target_test, prediction)


optuna.logging.set_verbosity(optuna.logging.WARNING)
sampler = TPESampler(seed=1)
study = optuna.create_study(study_name="catboost", direction="minimize", sampler=sampler)
study.optimize(lambda trial: objective(trial, features_train, target_train_rougher, 
                                       features_test, target_test_rougher), n_trials=100)
trial_rougher = study.best_trial
print('Для целевого признака rougher.output.recovery:')
print(study.best_params)
for key, value in trial_rougher.params.items():
    print("    {}: {}".format(key, value))
    
study.optimize(lambda trial: objective(trial, features_train, target_train_final, 
                                       features_test, target_test_final), n_trials=100)
trial_final = study.best_trial
print('Для целевого признака final.output.recovery:')
print(study.best_params)
for key, value in trial_final.params.items():
    print("    {}: {}".format(key, value))
```
