**Стиль заголовков сделан в 'верблюжьем' регистре, переведем все в 'змеиный' регистр.**

Напишем функцию перевода:

```python
# функция получает на вход список из наименований, а также задан разделитель
def camel_to_snake(columns, sep='_'): 
    # новый список, который будет формироваться из новых наименований
    columns_new = [] 
    # переберем каждое слово списка
    for column in columns:
        # новое слово в змеином регистре
        snake_register = ''
        # перебираем каждое слово по букве
        for i in column:
            # если встретили символ в верхнем регистре
            if i.isupper(): 
                # добавим разделитель и этот символ в нижнем регистре
                snake_register += sep + i.lower()
            else:
                snake_register += i
        snake_register = snake_register.lstrip(sep)
        # добавим новое слово в новый список 
        columns_new.append(snake_register)
    return columns_new
```

### Дерево решений DecisionTreeClassifier

**Воспользуемся функцией `DecisionTreeClassifier` импортируем ее из модуля `sklearn.tree`.**

Проверим, какие наилучшие значения метрик `f1` и `auc_roc` получим при различных значениях гиперпараметра `max_depth` - максимальная глубина дерева и `min_samples_split` -  минимальное количество объектов в узле, `min_samples_leaf` -  минимальное количество объектов в листе. 

```python
def  decision_tree(target_train, features_train, 
                   target_valid, features_valid, class_weight=None):  
# в данные переменные будем записывать значения лучших метрик
    best_f1 = 0
    best_auc_roc = 0.5
# в данные списки будем записывать параметры лучших метрик
    best_f1_param = []
    best_auc_roc_param = []

    for min_samples_split in range(2, 5):
        for min_samples_leaf in range(2, 5):
            for max_depth in range(1, 10):
                # построение модели
                model_tree = DecisionTreeClassifier(random_state=12345, 
                                        max_depth=max_depth, 
                                        min_samples_split=min_samples_split, 
                                        min_samples_leaf=min_samples_leaf,
                                        class_weight=class_weight)
                # обучение модели на тренировочной выборке
                model_tree.fit(features_train, target_train) 
                # получим предсказания модели 
                predictions_tree = model_tree.predict(features_valid)
                # подсчет метрик модели на валидационной выборке
                f1 = f1_score(target_valid, predictions_tree) 
                auc_roc = roc_auc_score(target_valid, 
                             model_tree.predict_proba(features_valid)[:, 1]) 

                if f1 > best_f1:
                    # параметры для наилучшей модели
                    best_f1_param = [max_depth, min_samples_split, 
                                     min_samples_leaf]
                    # наилучшее значение метрики на валидационных данных
                    best_f1 = f1 

                if auc_roc > best_auc_roc:
                    # параметры для наилучшей модели
                    best_auc_roc_param = [max_depth, min_samples_split, 
                                          min_samples_leaf]
                    # наилучшее значение метрики на валидационных данных                
                    best_auc_roc = auc_roc 
    return best_f1, best_f1_param, best_auc_roc, best_auc_roc_param
```

### Cлучайный лес RandomForestClassifier

**Воспользуемся функцией `RandomForestClassifier` импортируем ее из модуля `sklearn.ensemble`.**

Проверим, какие наилучшие значения метрик `f1` и `auc_roc` получим при различных значениях гиперпараметра `n_estimators` - количество деревьев, `max_depth` - максимальная глубина дерева, `min_samples_split` -  минимальное количество объектов в узле и `min_samples_leaf` -  минимальное количество объектов в листе. 

```python
def  random_forest(target_train, features_train, 
                   target_valid, features_valid, class_weight=None):
    # в данные переменные будем записывать значения лучших метрик
    best_f1 = 0
    best_auc_roc = 0.5

    # в данные списки будем записывать параметры лучших метрик
    best_f1_param = []
    best_auc_roc_param = []

    for n_estimators in range(1, 30):
        for min_samples_split in range(2, 5):
            for min_samples_leaf in range(2, 5):
                for max_depth in range(1, 10):
                    # построение модели
                    model_forest = RandomForestClassifier(random_state=12345, 
                                        n_estimators=n_estimators,
                                        max_depth=max_depth, 
                                        min_samples_split=min_samples_split, 
                                        min_samples_leaf=min_samples_leaf,
                                        class_weight=class_weight)
                    # обучение модели на тренировочной выборке
                    model_forest.fit(features_train, target_train) 
                    # получим предсказания модели 
                    predictions_forest = model_forest.predict(features_valid)
                    # подсчет метрик модели на валидационной выборке
                    f1 = f1_score(target_valid, predictions_forest) 
                    auc_roc = roc_auc_score(target_valid, 
                            model_forest.predict_proba(features_valid)[:, 1]) 

                    if f1 > best_f1:
                        # параметры для наилучшей модели
                        best_f1_param = [n_estimators, max_depth, 
                                         min_samples_split, min_samples_leaf]
                        # наилучшее значение метрики на валидационных данных
                        best_f1 = f1 

                    if auc_roc > best_auc_roc:
                        # параметры для наилучшей модели
                        best_auc_roc_param = [n_estimators, max_depth, 
                                              min_samples_split, 
                                              min_samples_leaf]
                        # наилучшее значение метрики на валидационных данных                
                        best_auc_roc = auc_roc 
    return best_f1, best_f1_param, best_auc_roc, best_auc_roc_param
```

### Логистическая регрессия LogisticRegression

Проверим, какие наилучшие значения метрик `f1` и `auc_roc` получим при различных значениях гиперпараметра `max_iter` - количество деревьев и `solver` - параметр выбора алгоритма.

### Увеличение выборки объектов редкого класса

Напишем функцию для увеличения выборки объектов редкого класса

```python
def upsample(features, target, repeat):
    # Разделим выборку на отрицательные и положительные объекты
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    # Копируем несколько раз положительные объекты и совмещаем в одну таблицу
    features_upsampled = pd.concat([features_zeros] + 
                                   [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + 
                                 [target_ones] * repeat)

    # Перемешиваем данные
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
        
    return features_upsampled, target_upsampled
```

### Уменьшение выборки объектов частого класса

Напишем функцию для уменьшения выборки объектов частого класса

```python
def downsample(features, target, fraction):
    # Разделим выборку на отрицательные и положительные объекты
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    # Копируем несколько раз положительные объекты и совмещаем в одну таблицу
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction,
                                    random_state=12345)] + [features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, 
                                    random_state=12345)] + [target_ones])

    # Перемешиваем данные
    features_downsampled, target_downsampled = shuffle(
    features_downsampled, target_downsampled, random_state=12345)
        
    return features_downsampled, target_downsampled
```

Новые выборки с уменьшенным количеством объектов частого класса:

```python
features_train_downsampled, target_train_downsampled = downsample(
                                        features_train, target_train, 0.25)
```
